### 3- Approche récursive naïve

<div class="alert alert-success">
    
Pour mettre en évidence le problème d'un traitement récursif classique, nous n'allons pas utiliser les jeux de données de l'année dernière (appels récursifs trop nombreux), mais celui-ci : 
    
Nous pourrons reprendre nos jeux de données lorsque nos programmes seront au point.
    
<img src="Images/SAD_recursion.png" alt="Arbre de récursion" width=80% align=center>
    

    
Lorsque l'on observe l'arbre de récursion, on constate que plusieurs solutions sont calculées plusieurs fois. L'idée de Bellman est de mémoriser dans un tableau en mémoire (mémoïzation), le résultat des calculs et avant d'en exécuter un nouveau, vérifier s'il n'existe pas déjà en mémoire, auquel cas, on prend la valeur mémorisée plutôt que de la recalculer. On réalise un cache. Il s’agit d’une approche dynamique de type « top-down ».
    
**Voici un exemple de traitement récursif**:
</div>

```python
def sac_a_dos_recursif_naif(info_objets,masse_maxi,nb):
    if nb==0:
        return 0
    if info_objets[nb-1][1]>masse_maxi:
        return sac_a_dos_recursif_naif(info_objets,masse_maxi,nb-1)
    else:
        return max(sac_a_dos_recursif_naif(info_objets,masse_maxi,nb-1),
                   info_objets[nb-1][2]+sac_a_dos_recursif_naif(info_objets,
                masse_maxi-info_objets[nb-1][1],nb-1))


#Structure de données
objets=['objet1','objet2','objet3','objet4']
masses=[3,2,1,1]
valeur=[5,3,4,3]
masse_maxi=5
nb_objets=len(objets)
ratio=[]
info_objets=[]
mon_container=[]
table_valeur=[[0 for x in range(masse_maxi+1)] for y in range(nb_objets+1)]

#Calcul du ratio Valeur/Poids
for i in range(nb_objets):
    ratio.append(valeur[i]/masses[i])

#Création d'une liste de tuples concernant les objets
for i in range(nb_objets):
    info_objets.append((objets[i],masses[i],valeur[i],ratio[i]))

#Tri de ratio dans l'ordre croissant (utilités)
info_objets.sort(key=lambda L:L[3], reverse=True)

mon_container=sac_a_dos_recursif_naif(info_objets,masse_maxi,nb_objets)

print(f'La valeur de mon container est de {mon_container}')
```

```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Algorithme : prendre la valeur maximale entre :
    - valeur de l'objet précédent pour une masse donnée
    ou
    - valeur de l'objet + valeur de l'objet précédent
    pour une masse minorée de la masse de l'objet.
    
    Complexité temporelle : 2 boucles for imbriquées :
        1 - nb_objets répétitons
        2 - masse_maxi répétitions
        = O(nb_objets x masse_maxi)
        
    Complexité en espace : O(2 x nb_objets x masse_maxi
"""



def sac_a_dos_dynamique(masse_maxi, masses, valeurs, nb_objets):
    """
    Cette fonction détermine grâce à une programmation dynamique
    la meilleure utilité (valeur) à embarquer dans un container
    tout en ne dépassant pas la charge maximale autorisée.
    L'objet n'est pas divisible (non fractionnaire), il est pris
    ou non (0-1)
    Entrée :
        - La masse maximale du container de type entier positif;
        - La liste des masses des objets à embarquer (entiers positifs)
        - La liste des valeurs des objets dans le même ordre que la liste
        précédente (entiers positifs) ;
        - Le nombre d'objets postulant à l'embarquement(entier positif).
    Pré-conditions :
        - La masse maximale est supérieure à 0 ;
        - La somme des masses des objets est supérieure à la masse maximale ;
        - La masse de chaque objet est inférieure à la masse maximale.
    Sortie :
        - La masse des objets embarqués ;
        - La valeur des valeurs cumulées embarquées.
    Tests :
    >>> sac_a_dos_dynamique(200,[140,60,65,20],[65,25,25,30],4)
    (160, 95)
    >>> sac_a_dos_dynamique(200,[140,60,65,20],[25,5,15,10],4)
    (160, 35)
    >>> sac_a_dos_dynamique(5,[3,2,1],[5,3,4],3)
    (4, 9)
    
    #Mémo : laisser une ligne en dessous du test !
    """   
    assert type(masse_maxi)== int and masse_maxi >0, 'La masse maximale doit être de type entier positif'
    assert sum(masses)>masse_maxi, 'L\'exécution du programme ne sert à rien : embarquez tout !'
    assert [x for x in masses if x>masse_maxi]==[], 'Un des objets a une masse supérieure à la masse maximale autorisée'
    assert [x for x in masses if x<=0]==[], 'Les masses des objets doivent être positives'
    assert [x for x in valeurs if x<=0]==[], 'Les valeurs des objets doivent être positives'
    assert type(nb_objets)== int and nb_objets >0, 'Le nombre d\'objets doit être de type entier positif'
         
    
    # Itération sur les lignes (objets)
    for index in range(1,nb_objets+1):
        # Itération sur les colonnes (masses)
        for masse in range(1,masse_maxi+1):
            # Si la masse de l'objet > à la valeur de la colonne
            # prendre la valeur au dessus, la solution a déjà été trouvée
            if masses[index-1]> masse:
                table_valeur[index][masse]=table_valeur[index-1][masse]
            else:
                # Si la masse de l'objet < à la valeur de la colonne
                valeur_precedente=table_valeur[index-1][masse]
                # valeur de l'objet courant + le la valeur du reste de masse
                valeur_nouvelle=valeurs[index-1]+table_valeur[index-1][masse-masses[index-1]]
                table_valeur[index][masse]=max(valeur_precedente,valeur_nouvelle)
                if valeur_nouvelle>=valeur_precedente:
                    table_choix[index][masse]=1
    return table_valeur[nb_objets][masse_maxi]

def chargement_container(nb):
    # chargement du container
    masse_disponible=masse_maxi
    masse_embarquee=0
    for index in range((nb_objets),0,-1):
        if table_choix[index][masse_disponible]:
            mon_container.append(objets[index-1])
            masse_disponible-=masses[index-1]
    masse_embarquee=masse_maxi-masse_disponible
    return masse_embarquee


def affichage_tables(table_valeur,table_choix):
    # présentation des résultats
    print('Table des valeurs')
    print('      Masse du container')
    print('----->',end="|")
    for index in range(masse_maxi+1):
        print(index, end="|")
    print('\n')  
    for index in range(1,nb_objets+1):
        print(objets[index-1],end="|")
        for masse in range(masse_maxi+1):
            print(table_valeur[index][masse], end="|")
        print('\n')  
        
    print('\nTable des choix')
    print('      Masse du container')
    print('----->',end="|")
    for index in range(masse_maxi+1):
        print(index, end="|")
    print('\n') 
    for index in range(1,nb_objets+1):
        print(objets[index-1],end="|")
        for masse in range(masse_maxi+1):
            print(table_choix[index][masse], end="|")
        print('\n') 



#Structure de données
objets=['objet1','objet2','objet3','objet4']
masses=[3,2,1,1]
valeurs=[5,3,4,3]
masse_maxi=5
nb_objets=len(objets)
mon_container=[]
masse_container=0
valeur_container=0
# Création des tables
table_valeur=[[0 for x in range(masse_maxi+1)] for y in range(nb_objets+1)]
table_choix=[[0 for x in range(masse_maxi+1)] for y in range(nb_objets+1)]

valeur_container=sac_a_dos_dynamique(masse_maxi, masses, valeurs, nb_objets)
masse_container=chargement_container(nb_objets)
print('Je vous propose d\'emporter avec vous : ')
for i in mon_container:
    print('- ',i)
print(f'La masse embarquée est de {masse_container} kg sur les {masse_maxi} kg autorisés, ce qui représente une valeur de {valeur_container}.')
```

```python
# Bellman Ford Algorithm in Python
# https://www.programiz.com/dsa/bellman-ford-algorithm


class Graph:

    def __init__(self, vertices):
        self.V = vertices   # Total number of vertices in the graph
        self.graph = []     # Array of edges

    # Add edges
    def add_edge(self, s, d, w):
        self.graph.append([s, d, w])

    # Print the solution
    def print_solution(self, dist):
        print("Vertex Distance from Source")
        for i in range(self.V):
            print("{0}\t\t{1}".format(i, dist[i]))

    def bellman_ford(self, src):

        # Step 1: fill the distance array and predecessor array
        dist = [float("Inf")] * self.V
        # Mark the source vertex
        dist[src] = 0

        # Step 2: relax edges |V| - 1 times
        for _ in range(self.V - 1):
            for s, d, w in self.graph:
                if dist[s] != float("Inf") and dist[s] + w < dist[d]:
                    dist[d] = dist[s] + w

        # Step 3: detect negative cycle
        # if value changes then we have a negative cycle in the graph
        # and we cannot find the shortest distances
        for s, d, w in self.graph:
            if dist[s] != float("Inf") and dist[s] + w < dist[d]:
                print("Graph contains negative weight cycle")
                return

        # No negative weight cycle found!
        # Print the distance and predecessor array
        self.print_solution(dist)


g = Graph(5)
g.add_edge(0, 1, 5)
g.add_edge(0, 2, 4)
g.add_edge(1, 3, 3)
g.add_edge(2, 1, 6)
g.add_edge(3, 2, 2)

g.bellman_ford(0)
```
