### 1- Projet N°1 :  alignement de séquences

<div class="alert alert-success">
    
Problématique :
    
Lorsque vous saisissez un texte dans un traitement de texte, et que vous faites une erreur de frappe en écrivant chein au lieu de chien, celui-vous propose une liste de mots comme : chien, hein, chemin, chine ... en correction.
    
Le génome humain est codé avec quatre bases nucléiques : Adénine, Cytosine, Thymine, Guanine. On peut le représenter par des très longs mots sur l’alphabet {A, C, T, G}. Comment détecter qu'un personne atteinte de la COVID-19, est contaminée avec le variant anglais ?
    
*Quel rapport entre ces deux contextes ?* On veut connaître la distance entre 2 chaînes de caractères.
    
  
La distance d’édition (ou de Levenshtein) entre deux mots (séquences) A et B est la longueur de la plus courte suite de
transformations pour passer de A à B, avec les transformations suivantes :
- insertion d’une nouvelle lettre ;
- suppression d’une lettre ;
- remplacement d’une lettre par une autre (mutation).
    
Chaque transformation aura un coût de 1.
    
Une définition alternative serait la suivante : nombre minimal de désaccords dans un alignement de A et B.
    
Exemples :
    
Pour passer de chein à chien (on ne peut pas permuter deux caractères):
    
chein    suppression du i (ou du e)
    
chen     insertion du i (ou du e)
    
chien
    
2 transformations : distance = 2
    
nombre de désaccords : 2
    
chein
    
chien
    
Pour passer de chein à chemin :
    
chein    insertion du m
    
chemin
    
1 transformation : distance = 1
    
nombre de désaccords : 1
    
che-in
    
chemin
    
On pourrait continuer sur les autres mots proposés, ils ne sont jamais "loin" du mot souhaité.

La distance entre les deux séquences nucléïques suivantes est de 3.
    
-ACTACGAAC
    
GACTA-GA-C

**Principe :**

Pour obtenir un alignement entre les séquences A et B, il suffit de considérer les cas d'alignements suivants :
    
**Cas limites :**

soit m, la longueur de la chaîne A

et n, la longueur de la chaîne B

**Si m=0**

la chaîne A est vide. Pour obtenir la chaîne B, il faut insérer les n caractères de B.

A = ' '

B= 'abc'

**Si n=0**

La chaîne B est vide. Pour obtenir la chaîne B, il faut supprimer les m caractères de A

A= 'abc'

B= ' '

**Cas classiques :**

**Même caractère dans les 2 chaînes :**

A = 'sol'

B = 'sel'

Le caractère *l* est identique dans les 2 chaînes. On commence par l'extrémité des chaînes et l'on remonte vers la gauche. Le caractère  *l* de A est à la position j (= m ici), alors que le caractère  *l* de B est à la position i (= n ici).

Les 2 caractères correspondants, il n'y a rien à faire comme opération. On sépare alors le préfixe de chaque chaîne et l'on cherche le nombre minimum d'opérations à réaliser pour la conversion  de *so* en *se*.

i et j sont décrémentés de 1 : (i-1),(j-1). Le résultat Res(i,j) pour Res(3,3)= Res(2,2) + 0.  
    
**Caractères différents dans les 2 chaînes :**

A = 'abc'	avec c en position m

B = 'bcd'	et d en position n

Nous avons le choix entre 3 opérations : inserer, supprimer, remplacer.

**Cas de l'insertion :** 

On insère le caractère d dans la position m pour qu'il corresponde avec le caractère d en position n. On décrémente j de 1 pour pointer le c alors que i ne change pas  : (i, j-1). Il faut aligner abc avec bc (abc- avec bcd)

Res(3,3)=Res(3,2) + 1 (insertion)

**Cas de la suppression :**

On supprime le caractère c de A. On décrémente i pour pointer le b alors que j ne change pas : (i-1,j)

Res(3,3)=Res(2,3) + 1 (suppression)

**Cas du remplacement :**

On remplace le caractère c par le caractère d de manière à ce qu'il correspondent. On décrémente alors i et j . (i-1, j-1)

Res(3,3)=Res(2,2) + 1 (remplacement)

La solution optimale est de trouver le minimum de ces 3 cas.    
    

    
**Méthode :**
  
Pour former un alignement entre deux séquences, A de longueur m et B de longueur n, on construit progressivement une matrice dans laquelle la case de coordonnées (i,j) contient la distance entre le préfixe de A et celui de B. La case (n, m) de la matrice contient donc la distance entre A et B. 
    
Si on note M la matrice, cela donne :
    
$$
M(i,j) = min\left\{
    \begin{array}{ll}
        M(i-1,j-1) + 1 & \mbox{si } A[i] ≠ B[j],& \mbox{sinon }M(i-1,j-1) + 0\\
        M(i,j-1) + 1\\
        M(i-1,j) + 1\\
    \end{array}
\right.
$$
    
On se sert de cette équation de récurrence pour remplir, ligne par ligne, de la gauche vers la droite, chaque case (i, j) avec i > 0 et j > 0. La matrice de programmation dynamique est de dimension n + 1 par m + 1, la numérotation des lignes et des colonnes commence à 0. L’initialisation de la matrice se fait de la manière suivante :
    
M(0,0)=0

M(i,0)=i∗1 (coût de 1)
    
M(0,j)=j∗1 (coût de 1).
    
    
    
Exemple :

|      | «  »  |   C   |   H   |   I   |   E   |   N   |
| :--: | :---: | :---: | :---: | :---: | :---: | :---: |
| «  » | **0** |   1   |   2   |   3   |   4   |   5   |
|  N   | **1** |   1   |   2   |   3   |   4   |   4   |
|  I   | **2** |   2   |   2   |   2   |   3   |   4   |
|  C   |   3   | **2** |   3   |   3   |   3   |   4   |
|  H   |   4   |   3   | **2** | **3** |   4   |   4   |
|  E   |   5   |   4   |   3   |   3   | **3** | **4** |

«  » : chaîne de caractère vide
    
<br>
    
**La distance entre les 2 chaînes de caractères se lit dans la cellule (n,m). Elle vaut 4 : 2 insertions, 2 suppressions)**
