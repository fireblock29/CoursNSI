```python
# (C) 2020 - ChenNux, Fireblock & Averell.exe
# Programme réalisé en Python 3 dans le cadre de la séquence 1 de NSI 1ère
# 
# 2-DcoDer est un programme de génération et de décryptage de 2dcodes simples utilisant la table ISO-8859.1 
# Une ligne par caractère et une colonne par digit binaire
# Ce programme utilise la bibliothèque PIL pour générer les images au format jpeg
#
# MESSAGE IMPORTANT : Pour connaître les commandes disponibles, entrez "help" ou "h" dans la console

from PIL import Image, ImageDraw
import time as tm
bdd=[]

def userHelp():
    print("  2-DcoDeR list of commands : \n"
            "  - draw : lance la procédure de génération de 2dcode\n"
            "  - decode : lance la procédure de décodage de 2dcode\n"
            "  - exit : quitte le programme\n"
            "  - credits : affiche les crédits\n")
    
def credits():
    print("  (C) 2020 - ChenNux, Fireblock & Averell.exe\n"
            "  Programme réalisé en Python 3 dans le cadre de la séquence 1 de NSI 1ère\n")
    
def get2dcodeContent(fileName):
    name = ""
    currentLetterBin = ""
    charID = 0
    try:
        im = Image.open(fileName)
    except FileNotFoundError:
        return "ERROR"
    px = im.load()
    
    tailleX, tailleY = im.width, im.height
    tailleCarre = int((im.width-1)/8)
    center = tailleCarre/2
    nbrX = 8
    nbrY = int((tailleY-1)/tailleCarre)
    
    for y in range(0, nbrY):
        currentLetterBin = ""
        charID = 0
        for x in range(0, nbrX):
            if((px[x*tailleCarre+center, y*tailleCarre+center])[0] != 255):
                charID += 2**(nbrX-x-1)
        name += chr(charID)
    
    return name
        
    
def maj(texte):
    texte_modifie=texte[0].upper()
    longueur = len(texte)
    for i in range(1, longueur):
        if  texte[i-1]==" ":
            #assert ord(texte[i]) <=122,"Caractère non traité"
            #assert ord(texte[i]) >= 65,"Caractère non traité"
            if ord(texte[i]) in range(65,96):
                car = texte[i]
            else:
                car = chr(ord(texte[i])-32)
            texte_modifie=texte_modifie+car
        else:
            texte_modifie=texte_modifie+texte[i]
    return texte_modifie

def strtobin(nom):
    liste = []
    for char in nom:
        mot = []
        ordchar = ord(char)
        assert ordchar <= 255, "Erreur : Carractère non compatible"
        for i in range(8):
            if ordchar % 2 == 1:
                mot.append("1")
            else:
                mot.append("0")
            ordchar //= 2
        liste.append(''.join(mot)[::-1])
    return liste


def draw2dcode(binList, taillePx, fileName):
    nombreCarresX = 8
    nombreCarresY = len(binList)
    
    xTaille = taillePx*nombreCarresX
    yTaille = taillePx*nombreCarresY
    
    im = Image.new("RGB", (xTaille+1, yTaille+1))
    img = ImageDraw.Draw(im)
    img.rectangle(((0,0),(xTaille, yTaille)), fill="#FFFFFF", outline="black")
    
    for x in range(0, nombreCarresY):
        for y in range(0, nombreCarresX):
            if((binList[x])[y]=='1'):
                fillColor = "#000000"
            else:
                fillColor = "#FFFFFF"
            img.rectangle(((taillePx*y, taillePx*x),(taillePx*(y+1), taillePx*(x+1))), fill=fillColor, outline="black")
                
    
    im.save(fileName+".jpg")

if __name__ == "__main__":
    
    print("Bienvenue sur 2-DcoDeR !")
    print("Astuce : entrez \"help\" pour voir la liste des commandes disponibles\n")
    
    reqList_help = ["h", "help", "Help", "H"]
    reqList_draw = ["Draw", "Gen", "gen", "draw", "g", "G"]
    reqList_decode = ["Decode", "decode", "d", "D"]
    reqList_credits = ["c", "Credits", "credits", "C"]
    reqList_quit = ["q", "Q", "e", "E", "quit", "Quit", "exit", "Exit"]
    reqList_fichierclient = ["fichier"]
    reqList_printbdd=['aff']
    req = ""
    
    while True:
        req = input("> user@2-DcoDeR : ")
        print("")
        if(req in reqList_help):
            userHelp()
        elif(req in reqList_credits):
            credits()
        elif(req in reqList_draw):
            txt = input("  Texte à générer : ")
            isMaj = input ("  Mise en forme auto ? (o/n) : ")
            taillePx = int(input("  Côté des carrés en px : "))
            filename = input("  Nom du fichier : ")
            startTime = tm.time()
            if(isMaj != "n"):
                draw2dcode(strtobin(maj(txt)), taillePx, filename)
            else:
                draw2dcode(strtobin(txt), taillePx, filename)
            endTime = tm.time()
            elapsedTime = endTime-startTime
            print("\n  2dcode correctement généré en "+str(round(elapsedTime, 3))+" secondes\n")
        elif(req in reqList_decode):
            file = input("  Nom du fichier : ")+".jpg"
            content = get2dcodeContent(file)
            if(content != "ERROR"):
                print("\n  Le contenu de ce 2Dcode est : "+content+"\n")
            else:
                print("\n  Le fichier "+file+" n'existe pas ou n'a pas pu être ouvert.\n")
        elif(req in reqList_quit):
            print("ChenNux, Averell et Fireblock vous souhaitent une bonne journée. A plus tard !")
            break
        else:
            print("  La commande \""+req+"\" n'existe pas. Entrez \"help\" pour la liste des commandes disponibles\n")
```
